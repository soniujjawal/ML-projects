```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 5))
df.set_index('Date')['Close'].plot(title='Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df.set_index('Date')['Close'], model='multiplicative', period=30)
result.plot();
```

```python
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(df['Close'].dropna())
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df['Close'].dropna(), lags=30)
plt.show()
plot_pacf(df['Close'].dropna(), lags=30)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(df['Close'], order=(5,1,0))
model_fit = model.fit()
print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=30)
plt.figure(figsize=(14,5))
df['Close'].plot(label='Historical')
forecast.index = pd.date_range(start=df['Date'].iloc[-1], periods=30, freq='D')
forecast.plot(label='Forecast', legend=True)
plt.grid()
plt.show()
```

```python
from prophet import Prophet
df_prophet = df[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})
prophet_model = Prophet()
prophet_model.fit(df_prophet)
future = prophet_model.make_future_dataframe(periods=30)
forecast = prophet_model.predict(future)
prophet_model.plot(forecast);
```

```python
# Preprocess for LSTM
import numpy as np
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df[['Close']])

# Create sequences
X, y = [], []
for i in range(60, len(data_scaled)):
    X.append(data_scaled[i-60:i])
    y.append(data_scaled[i])

X, y = np.array(X), np.array(y)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Build LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X, y, epochs=10, batch_size=32)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

def evaluate(true, pred):
    rmse = np.sqrt(mean_squared_error(true, pred))
    mae = mean_absolute_error(true, pred)
    print(f'RMSE: {rmse:.2f}, MAE: {mae:.2f}')
```
